```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("S&P500_Data.csv")
```

```python
data.isnull().sum()
```

```python
data.drop(columns='OldestDataDate',inplace=True)
```

```python
data['Underaged'] = np.where(data['Q4_NetIncome'].isna(), 1, 0)
```

```python
data['Underaged'].value_counts()
```

```python
data['in_index'] = 1
```

## EDA

```python
df = data
```

```python
bins = [0, 14.6e9, 50e9, 100e9, 500e9, 1e12, float('inf')]
labels = ['0-14.6B', '14.6B-50B', '50B-100B', '100B-500B', '500B-1T', '1T+']

df['Market Cap Category'] = pd.cut(df['Market Cap'], bins=bins, labels=labels)

category_counts = df['Market Cap Category'].value_counts().reindex(labels)

plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color='royalblue')
plt.title('Distribution of Market Cap')
plt.ylabel('Number of Stocks')
plt.xlabel('Market Cap Range')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
under_14_6B_companies = df[df['Market Cap Category'] == '0-14.6B']
print(under_14_6B_companies['Ticker'])
```

```python
f, ax = plt.subplots(figsize=(12, 9))

monthly_columns = [f"Month{i}_Volume" for i in range(1, 25)]
data[monthly_columns].mean().plot()
plt.title('Average Monthly Volume Over 24 Months')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(range(len(df['Q8_NetIncome'].dropna())), df['Q8_NetIncome'].dropna(), color='skyblue', edgecolor='black')
plt.title('Scatter Plot of Net Incomes for the Last Quarter')
plt.xlabel('Companies')
plt.ylabel('Net Income')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Outstanding Shares'].dropna(), bins=50, color='lightcoral', edgecolor='black')
plt.title('Distribution of Outstanding Shares')
plt.xlabel('Number of Shares')
plt.ylabel('Number of Companies')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
net_income_columns = [col for col in df.columns if "NetIncome" in col]
avg_net_income = df[net_income_columns].mean()

plt.figure(figsize=(12, 6))
avg_net_income.plot(kind='line', marker='o', color='mediumseagreen', linestyle='-')
plt.title('Average Net Income Over the Last 8 Quarters')
plt.ylabel('Average Net Income')
plt.xlabel('Quarters')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
df['Yearly_NetIncome'] = df['Q1_NetIncome'] + df['Q2_NetIncome'] + df['Q3_NetIncome'] + df['Q4_NetIncome']
df['Profitable'] = df['Yearly_NetIncome'].apply(lambda x: 'Profitable' if x > 0 else 'Not Profitable')

profitability_counts = df['Profitable'].value_counts()

plt.figure(figsize=(8, 6))
profitability_counts.plot(kind='bar', color=['green', 'red'])
plt.title('Yearly Profitability of Companies')
plt.ylabel('Number of Companies')
plt.xlabel('Profitability Status')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
not_profitable_companies = df[df['Profitable'] == 'Not Profitable']

print(not_profitable_companies['Ticker'])
```

```python
df['Avg_Volume_1_6'] = df[['Month1_Volume', 'Month2_Volume', 'Month3_Volume', 'Month4_Volume', 'Month5_Volume', 'Month6_Volume']].mean(axis=1)

df['Volume_Category'] = df['Avg_Volume_1_6'].apply(lambda x: 'Good Volume' if x > 250000 else 'Bad Volume')

plt.figure(figsize=(8, 6))
df['Volume_Category'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Distribution of Average Trading Volume (Month 1-6)')
plt.ylabel('Number of Companies')
plt.xlabel('Volume Category')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
bad_volume_companies = df[df['Volume_Category'] == 'Bad Volume']

print(bad_volume_companies['Ticker'])
```

```python
df['Avg Monthly Volume'] = df[['Month1_Volume', 'Month2_Volume', 'Month3_Volume', 'Month4_Volume', 'Month5_Volume', 'Month6_Volume']].mean(axis=1)

plt.figure(figsize=(12, 8))
plt.scatter(df['Avg Monthly Volume'], df['Market Cap'], color='blue', alpha=0.5)
plt.title('Market Cap vs. Average Monthly Volume')
plt.xlabel('Average Monthly Volume (over 6 months)')
plt.ylabel('Market Cap')
plt.yscale('log')
plt.xscale('log')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
bins = [0, 14.6e9, 50e9, 100e9, 500e9, 1e12, float('inf')]
labels = ['0-14.6B', '14.6B-50B', '50B-100B', '100B-500B', '500B-1T', '1T+']

df['Market Cap Category'] = pd.cut(df['Market Cap'], bins=bins, labels=labels)

colors = {'0-14.6B':'blue', '14.6B-50B':'cyan', '50B-100B':'green', '100B-500B':'yellow', '500B-1T':'orange', '1T+':'red'}

plt.figure(figsize=(15, 8))
for label in labels:
    subset = df[df['Market Cap Category'] == label]
    plt.scatter(subset.index, subset['Volume to Income Ratio'], s=100, 
                color=colors[label], edgecolor='black', label=label)
    
plt.title('Volume to Income Ratio for Each Company (Color-coded by Market Cap)')
plt.ylabel('Volume to Income Ratio')
plt.xlabel('Companies')
plt.xticks(ticks=[], labels=[])  
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
bins = [0, 14.6e9, 50e9, 100e9, 500e9, 1e12, float('inf')]
labels = ['0-14.6B', '14.6B-50B', '50B-100B', '100B-500B', '500B-1T', '1T+']

df['Market Cap Category'] = pd.cut(df['Market Cap'], bins=bins, labels=labels)

colors = {'0-14.6B':'blue', '14.6B-50B':'cyan', '50B-100B':'green', '100B-500B':'yellow', '500B-1T':'orange', '1T+':'red'}

plt.figure(figsize=(15, 8))
for label in labels:
    subset = df[df['Market Cap Category'] == label]
    plt.scatter(subset.index, subset['Market Cap to Avg Monthly Income Ratio'], s=100, 
                color=colors[label], edgecolor='black', label=label)
    
plt.title('Market Cap to Average Monthly Income Ratio for Each Company (Color-coded by Market Cap)')
plt.ylabel('Market Cap to Avg Monthly Income Ratio')
plt.xlabel('Companies')
plt.xticks(ticks=[], labels=[]) 
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
filtered_df = df[(df['Market Cap to Avg Monthly Income Ratio'] >= 0) & 
                 (df['Market Cap to Avg Monthly Income Ratio'] <= 3000)]

plt.figure(figsize=(15, 8))
for label in labels:
    subset = filtered_df[filtered_df['Market Cap Category'] == label]
    plt.scatter(subset.index, subset['Market Cap to Avg Monthly Income Ratio'], s=100, 
                color=colors[label], edgecolor='black', label=label)
    
plt.title('Market Cap to Average Monthly Income Ratio for Each Company (Color-coded by Market Cap)')
plt.ylabel('Market Cap to Avg Monthly Income Ratio')
plt.xlabel('Companies')
plt.xticks(ticks=[], labels=[]) 
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
avg_ratios = df.groupby('Market Cap Category')['Market Cap to Avg Monthly Income Ratio'].mean()

# Plot
plt.figure(figsize=(10, 6))
avg_ratios.plot(kind='bar', color='royalblue')
plt.title('Average Market Cap to Monthly Income Ratio by Market Cap Category')
plt.ylabel('Average Market Cap to Monthly Income Ratio')
plt.xlabel('Market Cap Range')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
avg_monthly_incomes = df.groupby('Market Cap Category')['Avg Monthly Income'].mean()

plt.figure(figsize=(10, 6))
avg_monthly_incomes.plot(kind='line', color='royalblue', marker='o')
plt.title('Average Monthly Income by Market Cap Category')
plt.ylabel('Average Monthly Income')
plt.xlabel('Market Cap Range')
plt.xticks(rotation=45)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
avg_monthly_volumes = df.groupby('Market Cap Category')['Avg Monthly Volume'].mean()

plt.figure(figsize=(10, 6))
avg_monthly_volumes.plot(kind='line', color='royalblue', marker='o')
plt.title('Average Monthly Trading Volume by Market Cap Category')
plt.ylabel('Average Monthly Trading Volume')
plt.xlabel('Market Cap Range')
plt.xticks(rotation=45)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
avg_net_incomes = df.filter(like="NetIncome").mean()

plt.figure(figsize=(15, 8))
avg_net_incomes.plot(kind='line', color='royalblue', marker='o')
plt.title('Average Net Income Over Different Quarters')
plt.ylabel('Average Net Income')
plt.xlabel('Quarters')
plt.xticks(rotation=45)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
avg_monthly_volumes = df.filter(like="Month").mean()

plt.figure(figsize=(15, 8))
avg_monthly_volumes.plot(kind='line', color='green', marker='o')
plt.title('Average Monthly Trading Volume Over Different Months')
plt.ylabel('Average Monthly Trading Volume')
plt.xlabel('Months')
plt.xticks(rotation=45)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
